```python
import json
```

```python
data_path = "C:\\projects\\pokefirered\\src\\data\\"
wild_data = json.load(open(data_path + "wild_encounters.json"))
```

```python
categories = [
    "land_mons",
    "water_mons",
    "rock_smash_mons",
    "fishing_mons"
]
```

```python
for map in wild_data["wild_encounter_groups"][0]["encounters"]:
    for category in categories:
        if category in map:
            for mon in map[category]["mons"]:
                if mon["species"] in devolution and devolution[mon["species"]] != mon["species"]:
                    print(f"changing {mon['species']} to {devolution[mon['species']]}")
                    mon["species"] = devolution[mon["species"]]
                    
    
```

```python
with open(data_path + "wild_encounters.json", "w") as f:
    json.dump(wild_data, f, indent=4)
```

```python
def check_pokemon():
    with open("baseline_pokemon.txt", "r") as f:
        baseline = f.read().split("\n")
    found = {}
    for p in baseline:
        found[p] = False
    for map in wild_data["wild_encounter_groups"][0]["encounters"]:
        for category in categories:
            if category in map:
                for mon in map[category]["mons"]:
                    if mon["species"] in found:
                        found[mon["species"]] = True
    print("missing pokemon:")
    count = 0
    for p in found:
        if not found[p]:
            count += 1
            print(p)
    print(f"total missing: {count}")
```

```python
check_pokemon()
```
