```python
import json
import os
```

```python
# This script will set the allow_running and allow_cycling flags to true for all maps in the maps directory.
maps_dir = "C:\\projects\\pokefirered\\data\\maps"
for root, dirs, files in os.walk(maps_dir):
    for file in files:
        if file == "map.json":
            fp = os.path.join(root, file)
            with open(fp, "r", encoding="utf8") as f:
                map_data = json.load(f)
            if "allow_running" in map_data:
                map_data["allow_running"] = True
            if "allow_cycling" in map_data:
                map_data["allow_cycling"] = True   
            with open(fp, "w", encoding="utf8") as f:
                json.dump(map_data, f, indent=4) 
    
```
