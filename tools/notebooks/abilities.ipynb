```python
from bs4 import BeautifulSoup
import requests
```

```python
template_token = "$ABILITY$"
url_template = f"https://bulbapedia.bulbagarden.net/wiki/{template_token}_(Ability)"
abilities = [
    "Motor Drive",
    "Multitype",
    "No Guard",
    "Normalize",
    "Poison Heal",
    "Quick Feet",
    "Reckless",
    "Rivalry",
    "Scrappy",
    "Simple",
    "Skill Link",
    "Slow Start",
    "Sniper",
    "Snow Cloak",
    "Snow Warning",
    "Solar Power",
    "Motor Drive",
    "Multitype",
    "No Guard",
    "Normalize",
    "Poison Heal",
    "Quick Feet",
    "Reckless",
    "Rivalry",
    "Scrappy",
    "Simple",
    "Skill Link",
    "Slow Start",
    "Sniper",
    "Snow Cloak",
    "Snow Warning",
    "Solar Power",
    "Solid Rock",
    "Stall",
    "Steadfast",
    "Storm Drain",
    "Super Luck",
    "Tangled Feet",
    "Technician",
    "Tinted Lens",
    "Unaware",
    "Unburden",
]

def format_ability(ability):
    return ability.replace(" ", "_")

def header_format_ability(ability):
    return f"ABILITY_{format_ability(ability).upper()}"

def get_ability_page(ability):
    url = url_template.replace(template_token, format_ability(ability))
    response = requests.get(url)
    return response.text

def is_int(string):
    try:
        int(string)
        return True
    except ValueError:
        return False
```

```python
def get_ability_pokemon(ability):
    page = get_ability_page(ability)
    soup = BeautifulSoup(page, 'html.parser')
    tables = soup.find_all('table')
    pokemon = {}

    # Iterate through tables to find the one with the desired header
    target_table = None
    for table in tables:
        # Find all header cells (th tags) in the table
        headers = table.find_all('th')
        for header in headers:
            if "First Ability" in header.get_text(strip=True):
                target_table = table
                break
        if target_table:
            break

    if target_table:
        # Extract rows from the target table
        rows = target_table.find_all('tr')
        table_data = []

        for row in rows:
            # Extract cell data from each row
            cells = row.find_all(['th', 'td'])
            row_data = [cell.get_text(strip=True) for cell in cells]
            table_data.append(row_data)

        # Print the extracted table data
        for row in table_data:
            if is_int(row[0]) and (ability in row[5] or ability in row[6]):
                p_name = row[2]
                pokemon[p_name] = (row[5], row[6])
    else:
        print("Table with header 'First Ability' not found.")
    return pokemon

```

```python
pokemon = {}
for ability in abilities:
    print(ability)
    abils = get_ability_pokemon(ability)
    for p in abils:
        pokemon[p] = []
        for a in abils[p]:
            formatted_ability = a.replace("Gen IV+", "")
            formatted_ability = formatted_ability.replace("Gen VII+", "")
            pokemon[p].append(formatted_ability)
print(pokemon)
```

```python
# reformat the dictionary so all keys are uppercase
pokemon = {k.upper(): v for k, v in pokemon.items()}
```
