```python
import os
```

```python
scripts_fp = "C:\\projects\\pokefirered"
file_exts = ["inc", "s"]
ignore_folders = ["agbcc", "asm", "constants", "src", "sound", "text"]
ignore_files = ["connections.inc", "events.inc", "header.inc", "text.inc"]
```

```python
# recursively walk through the directory
script_files = []
for root, dirs, files in os.walk(scripts_fp):
    for file in files:
        if any(ignore_folder in root for ignore_folder in ignore_folders):
            continue
        if file in ignore_files:
            continue
        if file.split(".")[-1] in file_exts:
            script_files.append(os.path.join(root, file))
```

```python
command_examples = {}
def extract_examples(script: str):
    lines = script.split("\n")
    for line in lines:
        if line.strip().endswith(":"):
            continue
        if line.strip().startswith("#") or line.strip().startswith("@") or line.strip().startswith("."):
            continue
        if line == "":
            continue
        if "::" in line or ":" in line:
            continue
        parts = line.strip().split(" ")
        if parts[0] == "":
            continue
        if parts[0] not in command_examples:
            command_examples[parts[0]] = line.strip()

```

```python
for script_file in script_files:
    with open(script_file, "r", encoding="utf8") as f:
        script = f.read()
        extract_examples(script)
```

```python
# alphabetize the commands
command_examples = dict(sorted(command_examples.items(), key=lambda item: item[0]))
```
